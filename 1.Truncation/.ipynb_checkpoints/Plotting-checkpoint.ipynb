```julia
function plot_fig_tau(trunc::Truncation,solution::AiyagariSolution,params::Params)
    
    @unpack α = params
    @unpack K,L = solution
    
    
    τs = collect(0.050:0.0005:0.090)
#     tau = collect(0.050:0.01:0.090)
    τs = collect(0.050:0.04:0.090)
    Y = K^α*L^(1-α)
    Ts = Y*τs

    dif_trunc = zeros(Float64, length(Ts))
    Sψ        = zeros(Float64, length(Ts))
    VT        = zeros(Float64, length(Ts)) #utility government good
    VTp       = zeros(Float64, length(Ts)) #marginal utility governmenmt good
    Sλ        = zeros(Float64, length(Ts))
    Rt        = zeros(Float64, length(Ts))
    wt        = zeros(Float64, length(Ts))
    Ybew      = zeros(Float64, length(Ts))
    tauoptbew = zeros(Float64, length(Ts))

    for (i,Tt) ∈ enumerate(Ts)
        params.Tt = -Tt #We change only the level of transfers
        @show Tt, params.Tt
        solution = steady(params);

        Ybew[i] = solution.K^α*solution.L^(1-α)
        tauoptbew[i] = Tt./Ybew[i]


        ##############################
        #   TRUNCATION METHOD
        ##############################

        trunc = Truncate(2,solution,params);


        ######Calculating the steady state
        #dif_truncation[i] = plan.CC[4]
        dif_trunc[i] = 0.0
        Sψ[i]  = sum(trunc.allocation_trunc.S_h.*trunc.lagrange_mult.ψh)
        VT[i]  = Tt.^(trunc.θ)
        VTp[i] = trunc.θ * (Tt.^(trunc.θ-1))
        Rt[i]  = solution.R
        wt[i]  = solution.w
        Sλ[i]  = 0.0#plan.CC[6]
        
    end

    tauopt = Ts./Ybew
    save("truncation.jld2", "dif_truncation", dif_trunc, "Sψ", Sψ ,
    "VT", VT, "VTp", VTp, "Rt", Rt, "wt", wt, "Sλ", Sλ, "tauopt_bew", tauopt)

    value, ind = findmin(abs.(Sψ.- VTp))

    p1= plot(tauopt, Sψ ,label="Sψ", linewidth = 2, dpi=200, color = :blue)
    p1= plot!(tauopt, VTp ,label="V'(T)", linewidth = 2, dpi=200,linestyle =:dash)

    Ucbew = Sψ.+ Sλ

    p1 = plot!(tauopt, Ucbew ,label="Uc", linewidth = 2, dpi=200,linestyle =:dash)
    xlabel!("τ")
    display(p1)
    savefig(p1, "Optimaltaxes.png")

end
```
